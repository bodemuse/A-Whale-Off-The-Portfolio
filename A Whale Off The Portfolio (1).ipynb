```python
# Initial imports
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path

%matplotlib inline
```

```python
# Whale Returns
whale_returns =pd.read_csv('whale_returns.csv')
whale_returns.dropna(inplace =True) #drop empty cells
whale_returns['Date'] = pd.to_datetime(whale_returns['Date']) #convert the dates to a DateTimeIndex
f_whale_returns= whale_returns.set_index('Date')


# Algorithmic Daily Returns
algo_returns =pd.read_csv('algo_returns.csv')
algo_returns.dropna(inplace =True) #drop empty cells
algo_returns['Date'] = pd.to_datetime(algo_returns['Date'])   #convert the dates to a DateTimeIndex
f_algo_returns = algo_returns.set_index('Date')


#S&P 500 Returns
sp500_history = pd.read_csv('sp500_history.csv')
sp500_history.dropna(inplace =True)
sp500_history['Date'] = pd.to_datetime(sp500_history['Date']) #convert the dates to a DateTimeIndex

#remove $ and convert values in SP_500 column to float
sp500_history['Close'] = sp500_history['Close'].str.replace('$','')
sp500_history['Close'] = sp500_history['Close'].astype('float')

#Convert the S&P 500 closing prices to SP 500 daily returns.
sp500_history = sp500_history.rename(columns = {'Close': 'SP 500'})
sp500_history = sp500_history.set_index('Date')

#Calculates percentage change in SP 500 returns
f_sp500_history = sp500_history.pct_change()

```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
combined_portfolio = pd.concat([f_whale_returns,f_algo_returns, f_sp500_history], axis = 1, join ='inner')
```

## Calculate and Plot the daily returns

```python
# Plot daily returns of all portfolios
combined_portfolio.dropna(inplace =True)
combined_portfolio
```

```python
#plot daily returns
combined_portfolio.plot.bar(figsize= (15,10))
```

##  Calculate and Plot cumulative returns

```python
#calculate cumulative returns
cumulative_returns = (1 + combined_portfolio).cumprod()
cumulative_returns.head()
```

```python
cumulative_returns.plot(figsize =(15,10))
```

## Calculate portfolio Standard Deviation

```python
combined_std = combined_portfolio.std()
combined_std
```

## Determine which portfolios are riskier than the S&P 500

```python
#converts the series to dictionary items
combined_portfolio_dict = combined_std.to_dict() 

#outputs portfolios with standard deviation > than the std of SP500
risky_portfolios = [x for x,y in combined_portfolio_dict.items() if y > 0.008587]
print(*risky_portfolios, sep =', ')
```

## Annualised Standard Deviation

```python
# Calculate the annualized standard deviation (252 trading days)
annualised_std = combined_std * np.sqrt(252)
annualised_std
```


## Sharpe Ratio

```python
#risk-free rate of return assuemed to be zero
sharpe_ratios = ((combined_portfolio.mean() - 0) * 252) / (combined_portfolio.std() * np.sqrt(252))
sharpe_ratios
```

# Custom Portfolio

### Read CSVs

```python
#Read CSV files

aapl = pd.read_csv("aapl_historical.csv", index_col="Trade DATE", infer_datetime_format=True, parse_dates=True)
aapl.sort_index(inplace=True)
goog = pd.read_csv('goog_historical.csv', index_col="Trade DATE", infer_datetime_format=True, parse_dates=True)
goog.sort_index(inplace=True)
costco = pd.read_csv('cost_historical.csv', index_col="Trade DATE", infer_datetime_format=True, parse_dates=True)
costco.sort_index(inplace=True)
```

```python
#combine stocks into one portfolio
custom_portfolio = pd.concat([aapl, goog, costco], axis = 0, join = 'inner')

#reset date index
custom_portfolio = custom_portfolio.reset_index()

# Reorganize portfolio data by having a column per symbol
custom_portfolio = custom_portfolio.pivot_table(values="NOCP", index="Trade DATE", columns="Symbol")
custom_portfolio
```

```python
# Calculate Daily Returns
custom_portfolio = custom_portfolio.pct_change()
custom_portfolio.head()
```

```python
#dropna's

custom_portfolio.dropna(inplace = True)
custom_portfolio
```

### Calculate Portfolio Returns with an equal amount of each stock

```python
# Calculate Portfolio weighted Returns with an equal amount of each stock
aapl_weight = 0.33
goog_weight = 0.33
costco_weight = 0.33

portfolio_returns = aapl_weight * custom_portfolio["AAPL"] + goog_weight * custom_portfolio["GOOG"] + costco_weight * custom_portfolio["COST"]
portfolio_returns.head()
```

### Adding portfolio returns to the DataFrame with the other portfolios.

```python
#Adding portfolio returns to the DataFrame with the other portfolios.
portfolio_and_returns = pd.concat([custom_portfolio, portfolio_returns], axis = 1, join ='inner')

#Renaming the fifth column to 'Weighted returns'
portfolio_and_returns.rename(columns = {0 : "WEIGHTED RETURNS"}, inplace = True)
portfolio_and_returns
```

```python
#calculate the percentatge change
p_and_r = portfolio_and_returns.pct_change()
p_and_r.dropna(inplace = True)
p_and_r
```

```python
#standard deviation of the combined portfolio
par_portfolio = p_and_r.std()
```

### Annualised Standard Deviation

```python
# Calculate the annualized standard deviation (252 trading days)
annualised_std = par_portfolio * np.sqrt(252)
annualised_std
```

### Rolling Standard Deviation 

```python
#calculates and plot 21-days window rolling statistics
p_and_r.rolling(window=21).mean().plot(figsize =(15,10))
```

### Calculate and plot correlation

```python
par_portfolio_corr = p_and_r.corr()
par_portfolio_corr
```

```python
import seaborn as sns
sns.heatmap(par_portfolio_corr, vmin=-1, vmax=1, center=0,  cmap="YlGnBu", linewidths=.5)
```
